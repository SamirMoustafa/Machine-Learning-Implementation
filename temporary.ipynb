```python
%matplotlib notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Plot Normal Distribution

```python
# Generate random sample with n=100000 data with mean = 0 and sd=1
mu = 0
sd = 1
n = 10000000
```

## Univariate Gaussian Distribution

```python
random_data = np.random.normal(mu, sd, n)
# Round data to the first dec.
random_data = np.round(random_data, 1)
```

```python
df = pd.DataFrame(data=random_data, columns=['Data'])
df.head()
```

```python
df['Data'].value_counts().sort_index().plot(kind='line', figsize=(9,4))
```

## Bivariate Gaussian Distribution

```python
n = 1000
random_data = np.random.normal(mu, sd, [n,2])
```

```python
df = pd.DataFrame(data=random_data, columns=['X','Y'])
df.head()
```

```python
df.plot(kind='scatter',x='X', y='Y', figsize=(5,5))
```
