### Create Random Data

```python
center_1 = np.array([1,1])
center_2 = np.array([5,5])
center_3 = np.array([8,1])
center_4 = np.array([10,5])

# Generate random data and center it to the three centers
data_1 = np.random.randn(200,2) + center_1
data_2 = np.random.randn(200,2) + center_2
data_3 = np.random.randn(200,2) + center_3
data_4 = np.random.randn(200,2) + center_4

data = {0:data_1, 
        1:data_2,
        2:data_3,
        3:data_4}
```

```python
p = [7.75,4]
p = np.array(p)
```

```python
plot_data_and_point(data, p)
```

### Algorithm

```python
def KNN(points,p,k=3): 
    distance=[] 
    for group in points.keys(): 
        for feature in points[group]: 
            #calculate the euclidean distance of p from training points  
            euclidean_distance = np.linalg.norm(feature-p)
            
            # Add a touple of form (group, distance) in the distance list 
            distance.append((group, euclidean_distance)) 

    # sort the distance list in ascending order 
    # and select first k distances 
    distance.sort(key=lambda x:x[1])
  
    return distance[:k]
```
