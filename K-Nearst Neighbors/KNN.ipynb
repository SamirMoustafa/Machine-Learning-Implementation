```python
import math
import numpy as np
from matplotlib import pyplot as plt
```

```python
center_1 = np.array([1,1])
center_2 = np.array([5,5])
center_3 = np.array([8,1])
center_4 = np.array([10,5])

# Generate random data and center it to the three centers
data_1 = np.random.randn(200,2) + center_1
data_2 = np.random.randn(200,2) + center_2
data_3 = np.random.randn(200,2) + center_3
data_4 = np.random.randn(200,2) + center_4

data = {0:data_1, 
        1:data_2,
        2:data_3,
        3:data_4}
```

```python
ploted_data = []
for i in data.keys():
    for j in data[i]:
        ploted_data.append(j)
ploted_data = np.array(ploted_data)
```

```python
p = [6,2.5]
p = np.array(p)
```

```python
plt.scatter(ploted_data[:,0], ploted_data[:,1], s=7)
for i in data.keys():
    x_mean = data[i][:,0].mean()
    y_mean = data[i][:,1].mean()
    plt.text(x_mean, y_mean, i, style='italic', bbox={'facecolor': 'red', 'alpha': 0.9, 'pad': 10})

# Plot the target data
plt.scatter(p[0], p[1], marker='o', c='y', s=150)
```

```python
def classify_point(points,p,k=3): 
    distance=[] 
    for group in points.keys(): 
        for feature in points[group]: 
            #calculate the euclidean distance of p from training points  
            euclidean_distance = np.linalg.norm(feature-p)
            
            # Add a touple of form (distance,group) in the distance list 
            distance.append((group, euclidean_distance)) 

    # sort the distance list in ascending order 
    # and select first k distances 
    distance.sort(key=lambda x:x[1])
  
    return distance[:k]
```

```python
k = 5

classify_point(data,p,k)
```
